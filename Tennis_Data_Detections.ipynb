```python
import cv2
import matplotlib.pyplot as plt

image_path = '/content/drive/MyDrive/Player-Stick.v1i.yolov8/test/images/videoplayback_mp4-0_jpg.rf.d5fd551309cd9013088f82e3e539f66d.jpg'  # Replace with your image path
img = cv2.imread(image_path)

results = model(img)

annotated_img = results[0].plot()

plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python



model = YOLO('/content/runs/detect/train/weights/best.pt')

video_path = '/tennis.mp4'
cap = cv2.VideoCapture(video_path)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/output_video.mp4', fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    annotated_frame = results[0].plot()

    out.write(annotated_frame)

cap.release()
out.release()

print("Detection complete. Video saved as output_video.mp4.")
```
